```python
df = pd.read_csv('/Users/skareti/Desktop/Independent Study/NYTRestaurantReviews/data/Data.csv', index_col=0)
df.drop_duplicates(inplace = True)
df.info()
```

```python
df100 = pd.read_csv('/Users/skareti/Desktop/Independent Study/NYTRestaurantReviews/data/NYT_Top_100_Restaurants_2024_with_Rank.csv')
df100.head()
```

```python
rename_mapping = {
    "Tatiana by Kwame Onwuachi": "Tatiana",
    "Torrisi": "Torrisi Bar & Restaurant",
    "Casa Mono and Bar Jamón": "Casa Mono",
    "Hainanese Chicken House": "Hainan Chicken House",
    "M. Wells": "M. Wells Steakhouse",
    "Caleta 111": "Caleta 111 Cevicheria",
    "Randazzo’s Clam Bar": "Randazzo's Clam Bar"  # Ensuring consistent quotes
}

df100['Restaurant Name'] = df100['Restaurant Name'].replace(rename_mapping)

# Perform a left join to identify missing entries
merged_df = df100.merge(right = df, left_on = 'Restaurant Name', right_on = 'Restaurant_name', how='left', indicator=True)
# Identify restaurants in the top 100 list that are not in the master dataset
missing_restaurants = merged_df[merged_df['_merge'] == 'left_only']

# # Display the missing restaurants
print(missing_restaurants.shape)
missing_restaurants
```

```python
missing_restaurants.to_csv('/Users/skareti/Desktop/Independent Study/NYTRestaurantReviews/data/missingtop100_manuallycode.csv')
```

```python
missing_restaurants_filled = pd.read_csv('/Users/skareti/Desktop/Independent Study/NYTRestaurantReviews/data/missingtop100_filled.csv')
missing_restaurants_filled
```

```python
df_amend = pd.concat([df, missing_restaurants_filled])
print(df_amend.shape)
df_amend.head()
```

```python
df_amend.to_csv('/Users/skareti/Desktop/Independent Study/NYTRestaurantReviews/data/df_manuallyfilled_restaurants.csv')
```

```python
#rename map2 for Ewe's Delicious Treats
rename_mapping2 = {
    "Ewe’s Delicious Treats": "Ewe's Delicious Treats"
}

df100['Restaurant Name'] = df100['Restaurant Name'].replace(rename_mapping2)

# Perform a left join to identify missing entries
merged_df = df_amend.merge(df100, left_on = 'Restaurant_name', right_on='Restaurant Name', how = 'left', indicator=True)
#sort by 'Restaurant_name' and 'Publishing_Date' to get most recent reviews
merged_df = merged_df.sort_values(by=['Restaurant_name', 'Publishing_Date'], ascending=[True, False])
# Create a new DataFrame with the most recent reviews
most_recent_reviews = merged_df.drop_duplicates(subset=['Restaurant_name'], keep='first')

# Add 'top100_2024' column: 1 if in the top 100 list, 0 otherwise
merged_df['top100_2024'] = 0
merged_df['Rank'] = 0

# Update the original merged_df with the indicators from the most recent reviews
most_recent_reviews['top100_2024'] = most_recent_reviews['Rank'].notna().astype(int)
most_recent_reviews['Rank'] = most_recent_reviews['Rank'].fillna(0).astype(int)

# Assign the indicator and Rank to the merged_df
merged_df.update(most_recent_reviews[['Restaurant_name', 'top100_2024', 'Rank']])

# Add the indicator and Rank only for the top 100 restaurants
merged_df.loc[most_recent_reviews.index, 'top100_2024'] = most_recent_reviews['Rank'].apply(lambda x: 1 if x > 0 else 0)
merged_df.loc[most_recent_reviews.index, 'Rank'] = most_recent_reviews['Rank']

# Display the final DataFrame
print(merged_df)

# Check the number of rows with top100_2024 == 1
print(merged_df[merged_df['top100_2024'] == 1].shape[0])
```

```python
merged_df.drop(columns = ['_merge', 'Restaurant Name'], inplace = True)
```
